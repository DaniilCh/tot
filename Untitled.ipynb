```python
import xgboost as xgb
from xgboost import XGBClassifier
import numpy as np
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import argparse


drugs = ['sd-101']
phases = ['1', '2']

classifier = XGBClassifier()
classifier.load_model('001.model')

data = pd.read_pickle('util_data.pkl')
data.head(10)
```

```python
drugs
```

```python
data['Interventions2'] = [drugs]*len(data)
data['Phases2'] = [phases]*len(data)
data.head(10)
```

```python
drugs_binarizer = MultiLabelBinarizer()
phase_binarizer = MultiLabelBinarizer()
```

```python
#data_new = pd.DataFrame(data['NCT Number'])
data_new = pd.DataFrame(phase_binarizer.fit_transform(data.Phases), columns = 'Is 1st phase ' + phase_binarizer.classes_, index = data.index)
data_new.head(10)
```

```python
data_new = data_new.join(pd.DataFrame(drugs_binarizer.fit_transform(data.Interventions), columns = 'Is 1st ' + drugs_binarizer.classes_, index = data.index))
data_new = data_new.join(pd.DataFrame(phase_binarizer.transform(data.Phases2), columns = 'Is 2nd phase  ' + phase_binarizer.classes_, index = data.index))
data_new = data_new.join(pd.DataFrame(drugs_binarizer.transform(data.Interventions2), columns = 'Is 2nd ' + drugs_binarizer.classes_, index = data.index))

classifier._le = LabelEncoder().fit([0, 1])
data['Is_similar'] = classifier.predict(np.array((data_new)))
res = []

for i in data.index:
    if data['Is_similar'][i] == 1:
        res.append(data['NCT Number'][i])

print(res)
```
